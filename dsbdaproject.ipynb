```python
import re

likes = []

for val in df['dish_liked'].dropna():
    array_split = re.split(',', val)
    for item in array_split:
        likes.append(item.strip())   # remove extra spaces
```

```python
favourite_food = pd.Series(likes).value_counts()
favourite_food.head(30)
```

```python
cmap = plt.get_cmap('Set3')

top_food = favourite_food.nlargest(n=20, keep='first')
color = cmap(np.arange(len(top_food)))

ax = top_food.plot(
    kind='bar',
    figsize=(18, 10),
    title='Top 20 Favourite Food Counts',
    color=color
)

# Add labels on bars
for i in ax.patches:
    ax.annotate(
        str(int(i.get_height())),
        (i.get_x() + i.get_width()/2, i.get_height()),
        ha='center',
        va='bottom'
    )

plt.xlabel('Food Items')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
v = df['cuisines'].value_counts()[:15]

plt.figure(figsize=(20, 8))

ax = sns.barplot(x=v.index, y=v.values, palette='Paired')

# Add annotations
for i in ax.patches:
    ax.annotate(
        str(int(i.get_height())),
        (i.get_x() + i.get_width()/2, i.get_height()),
        ha='center',
        va='bottom'
    )

plt.title('Most Popular Cuisines of Bangalore', fontsize=20)
plt.xlabel('Cuisines', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
